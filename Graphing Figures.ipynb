## Graphing Figures

Lets look at making pretty figures. Lets say you have the following data

```python
x= [1,4,9,16,25]
y= [1,2,3,4,5]
```

We can bring in the matplotlib library (which you have hopfully had a chance to look at) and plot figures

```python
#improting matplotlib library
import matplotlib.pyplot as plt

plt.plot(x,y)
plt.show()
```

This figure is okay but we usually don't want to connect our data points with straight lines. A scatter plot would be better. 

```python
plt.scatter(x,y)
plt.show()
```

Thats better but lets look at some additional features we can use to make our figure a bit better. If we have an additional data set it could be hard to tell the two appart.

```python
x= [1,4,9,16,25]
y_1 = [1,2,3,4,5]
y_2= [0,1.38,2.2,2.77,3.22]
plt.scatter(x, y_1, c= "b")
plt.scatter(x, y_2, c="b")
plt.show()
```

matplotlib comes with a bunch of features that we can use to customize our figures.

```python
'''
scatter(x values, 
        y values, 
        s = size of maker, 
        c = marker color, 
        marker = marker style, 
        edgecolors = marker outline color,
        linewidths = width outline lines )
'''

plt.scatter(x, y_1, c= "b", marker = "^")
plt.scatter(x, y_2, c="r", marker = "o")
plt.show()

#https://matplotlib.org/3.2.1/api/markers_api.html
```

That's already looking better. Lets see what happens if we get even more fancy adding  edges and hexcode colors.

```python
plt.scatter(x, y_1, s=60 , c= "#990099", marker = "^" , edgecolors = "#000000")
plt.scatter(x, y_2, s=60 , c="#ff6600", marker = "o", edgecolors = "#000000")
plt.show()
#https://www.hexcolortool.com/
```

Okay maybe thats a bit too much but we can always change this one back. Lets add some text too

```python
plt.scatter(x, y_1, c= "b", marker = "^")
plt.scatter(x, y_2, c="r", marker = "o")

'''
xlabel(text, labelpad = off set from axis)
'''

plt.xlabel('x label', labelpad = 10)
plt.ylabel('y label', labelpad = 10)
plt.show()
```

And a ledgend!

```python
plt.scatter(x, y_1, c= "b", marker = "^", label = "data set 1")
plt.scatter(x, y_2, c="r", marker = "o", label = "data set 2")

plt.xlabel('x label', labelpad = 10)
plt.ylabel('y label', labelpad = 10)

plt.legend()
plt.show()
```

We can set the placement of the legend too.

```python
plt.scatter(x, y_1, c= "b", marker = "^", label = "data set 1")
plt.scatter(x, y_2, c="r", marker = "o", label = "data set 2")

plt.xlabel('x label', labelpad = 10)
plt.ylabel('y label', labelpad = 10)

plt.legend(loc='lower right',shadow=True, ncol=1)

plt.show()
```

```python
y_1_error = [0.3,0.2,0.5,0.2,0.4]
y_2_error = [0.2,0.5,0.4,0.2,0.3]

plt.scatter(x, y_1, c= "b", marker = "^", label = "data set 1")
plt.errorbar(x, y_1, yerr = y_1_error, ecolor = "b", capsize = 4, fmt= "none" )

plt.scatter(x, y_2, c="r", marker = "o", label = "data set 2")
plt.errorbar(x, y_2, yerr = y_2_error, ecolor = "r", capsize = 4, fmt= "none" )

plt.xlabel('x label', labelpad = 10)
plt.ylabel('y label', labelpad = 10)

plt.legend(loc='lower right',shadow=True, ncol=1)

plt.show()
```

We can even make things easier by scatter plotting with the error function

```python
#call help(plt.errorbar) to see specifics of using errorbars
plt.errorbar(x, y_1, yerr = y_1_error, ecolor = "b", capsize = 4, fmt= "b^", label = "data set 1")
plt.errorbar(x, y_2, yerr = y_2_error, ecolor = "r", capsize = 4, fmt= "ro", label =  "data set 2")

plt.xlabel('x label', labelpad = 10)
plt.ylabel('y label', labelpad = 10)

plt.legend(loc='lower right',shadow=True, ncol=1)

plt.show()
```

Lets format the graph too! First lets clean up those tick marks

```python
plt.errorbar(x, y_1, yerr = y_1_error, ecolor = "b", capsize = 4, fmt= "b^", label = "data set 1")
plt.errorbar(x, y_2, yerr = y_2_error, ecolor = "r", capsize = 4, fmt= "ro", label =  "data set 2")

plt.xlabel('x label', labelpad = 10)
plt.ylabel('y label', labelpad = 10)

plt.legend(loc='lower right',shadow=True, ncol=1)

#added a custom list to manual specify y tick locations
plt.yticks([0,2,4,6])
#this time I used the rage function to create a list of even numbers between 0 and 26
plt.xticks(list(range(0,27,2)))
#call help(plt.tick_params) to see specifics of using tick_params
plt.tick_params(axis = "both", direction = "in", length = 5, pad = 7)

plt.show()
```

We can also resize the figure by changing the figure size at the start of our code 

```python
plt.figure (figsize=(10,5))

plt.errorbar(x, y_1, yerr = y_1_error, ecolor = "b", capsize = 4, fmt= "b^", label = "data set 1")
plt.errorbar(x, y_2, yerr = y_2_error, ecolor = "r", capsize = 4, fmt= "ro", label =  "data set 2")

plt.xlabel('x label', labelpad = 10)
plt.ylabel('y label', labelpad = 10)

plt.legend(loc='upper left',shadow=True, ncol=1)

#added a custom list to manual specify y tick locations
plt.yticks([0,2,4,6])
#this time I used the rage function to create a list of even numbers between 0 and 26
plt.xticks(list(range(0,27,2)))
#call help(plt.tick_params) to see specifics of using tick_params
plt.tick_params(axis = "both", direction = "in", length = 5, pad = 7)

plt.show()
```

From where we started this is looking much better! :D
